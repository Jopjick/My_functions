```python
#считывание из файла
with open('files.txt', 'r') as file:
    files = [line.strip().split() for line in file.readlines()]
groups = []
for file in files:
    file[0] = file[0].split('.')
    
#объединение в группы    
while True:
    temp = files[0][0][1]
    group = []
    files1 = files.copy()
    for file in files1:
        if file[0][1] == temp:
            group.append(files.pop(files.index(file)))
    groups.append(group)
    if files == []:
        break
        
#обработка групп        
for group in groups:
    group.append(group[0][0][1])
    for i in range(len(group)-1):
        group[i][0] = group[i][0][0]+'.'+group[i][0][1]
        group[i][1] = int(group[i][1])
        name = group[i][2]
        if name == 'KB':
            group[i][1] = group[i][1]*1024
            group[i][2] = 'B'
        elif name == 'MB':
            group[i][1] = group[i][1]*1024*1024
            group[i][2] = 'B'
        elif name == 'GB':
            group[i][1] = group[i][1]*1024*1024*1024
            group[i][2] = 'B'
            
#подсчет веса файлов в группах
for group in groups:
    weight = 0
    for line in group[:-1]:
        weight += line[1]
    if weight > 1023:
        weight = round(weight / 1024)
        if weight > 1023:
            weight = round(weight / 1023)
            if weight > 1023:
                weight = round(weight / 1023)
                weight = str(weight)+' GB'
                group.append(weight)
            else:
                weight = str(weight)+' MB'
                group.append(weight)
        else:
            weight = str(weight)+' KB'
            group.append(weight)
    else:
        weight = str(weight)+' B'
        group.append(weight)
n = 0        
#сортированный вывод    
groups = sorted(groups, key=lambda x: x[-2])
for group in groups:
    n += 1
    temp = group[-1]
    group = sorted(group[:-2], key=lambda line: line[0])
    for line in group:
        print(line[0])
    print('----------')
    if n != len(groups):
        print('Summary: '+ temp)
        print()
    else:
        print('Summary: '+ temp, end='')
    
```
