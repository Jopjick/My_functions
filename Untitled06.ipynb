```python
import csv
with open('salary_data.csv', 'r', encoding='utf-8')as file:
    data = list(csv.DictReader(file, delimiter=';'))
for item in data:
    item['salary'] = int(item['salary'])
result = {}
for item in data:
    if item['company_name'] not in result.keys():
        result[item['company_name']] = [item['salary'], 1]
    else: result[item['company_name']] = [result[item['company_name']][0] + item['salary'], result[item['company_name']][1] + 1]  
for key in result:
    result[key] = result[key][0] / result[key][1]
result = sorted(result.items(), key=lambda x: x[1])
print(*map(lambda x: x[0], result), sep='\n')
```

```python
def get_columns(fil):
    import csv
    with open(fil, 'r', encoding='utf-8') as file:
        data = list(csv.reader(file))
    res = {}
    for i in range(len(data[0])):
        temp = [data[j][i] for j in range(len(data))]
        res[temp[0]] = temp[1:]
    return res
```

```python
import csv
with open('data.csv', 'r') as file:
    data = list(map(lambda x: x[2].split('@'), list(csv.reader(file))[1:]))
data, result = [item[1] for item in data], {}

for item in data:
    if item not in result.keys():
        result[item] = 1
    else:
        result[item] = result[item] + 1
result = sorted(list(map(lambda x: [x, result[x]], result)), key=lambda x: (x[1], x[0]))

with open('domain_usage.csv', 'w') as file:
    writer=csv.writer(file)
    writer.writerow(['domain', 'count'])
    for row in result:
        writer.writerow(row)
```

```python
import csv
with open('wifi.csv', 'r', encoding='utf-8') as source:
    data = list(csv.DictReader(source, delimiter=';'))
result = {}
for row in data:
    if row['district'] not in result.keys():
        result[row['district']] = int(row['number_of_access_points'])
    else:
        result[row['district']] = result[row['district']] + int(row['number_of_access_points'])
        
result = sorted(list(map(lambda x: [x, result[x]], result)))
result = sorted(result, key=lambda x: x[1], reverse=True)        
for item in result:
    print(*item, sep=': ')
```

```python
import csv
with open('titanic.csv', 'r', encoding='utf-8') as file:
    data = list(csv.DictReader(file, delimiter=';'))
data = list(filter(lambda x: x['survived'] == '1' and float(x['age']) < 18, data))
data = list(map(lambda row: [row['name'], row['sex']], data))
data = map(lambda x: x[0], sorted(data, key=lambda x: x[1], reverse=True))
print(*data, sep='\n')
```

```python
import csv
from datetime import datetime
with open('name_log.csv', 'r', encoding='utf-8') as file:
    d = list(csv.reader(file))
    col = d[0]
    data = [[row[0], row[1], datetime.strptime(row[2], '%d/%m/%Y %H:%M')] for row in d[1:]]
    
new_data = {}
for row in data:
    if row[1] not in new_data.keys():
        new_data[row[1]] = [row[0], row[2]]
    else:
        if new_data[row[1]][1] < row[2]:
            new_data[row[1]] = [row[0], row[2]]
            
new_data = list(map(lambda x: [new_data[x][0], x, new_data[x][1].strftime('%d/%m/%Y %H:%M')], new_data))
new_data = sorted(new_data, key=lambda x: x[1])

with open('new_name_log.csv', 'w', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(col)
    for row in new_data:
        writer.writerow(row)
```
