```python
def moment(data, av, k):
    return sum(map(lambda x: (x - av)**k, data))/len(data)

def medi(data):
    n = len(data)
    data1 = sorted(data) 
    if n % 2 == 1:
        return data1[n // 2]
    else:
        return (data1[n // 2] + data1[n // 2 + 1])/2
```

```python
with open('13_6', 'r') as file:
    data = file.readlines()
    
data = list(map(lambda x: [int(x[2]), int(x[3])], [item.strip().split('\t') for item in data]))
data_6, data_13, n = list(map(lambda x: x[0], data)), list(map(lambda x: x[1], data)), len(data)

av_13, av_6 = sum(data_13)/n, sum(data_6)/n
disp_13, disp_6 = moment(data_13, av_13, 2), moment(data_6, av_6, 2)
medi_13, medi_6 = medi(data_13), medi(data_6) 

print(av_13, medi_13, round(disp_13, 1), sep=', ')
```

```python
data_1 = [data_13[i] - data_6[i] for i in range(n)]

av_1 = sum(data_1)/n
exc = moment(data_1, av_1, 4)/(moment(data_1, av_1, 2))**2 - 3
ass = moment(data_1, av_1, 3)/(moment(data_1, av_1, 2))**1.5

print(round(av_1, 2), round(exc, 2), round(ass, 2), sep=', ')
```

```python
with open('colleges.txt', 'rt') as file:
    data_0 = [line.strip().split('\t') for line in file.readlines()]
keys, data = data_0.pop(0), []

for line in data_0: #все учебные заведения
    temp = {}
    for i in range(len(keys)):
        temp[keys[i]] = line[i]
    data.append(temp)
    
type_un, type_art = [], [] #группировка заведений по типам
for obj in data:
    if obj['School_Type'] == 'Univ':
        type_un.append(obj)
    else:
        type_art.append(obj)

data_un = {} #ключевая информация о заведениях
for key in keys[2:]:
    data_un[key] = list(map(lambda x: int(x[key]), type_un))

data_art = {}
for key in keys[2:]:
    data_art[key] = list(map(lambda x: int(x[key]), type_art))    
```

```python
av_SAT_art = sum(data_art['SAT'])/len(data_art['SAT'])

acc = sorted(data_un['Acceptance'])
Q3_acc = acc[int(len(acc)*3/4)]
print(acc, len(acc))
name = max(data, key=lambda x: int(x['$/Student']))['School']

print(round(av_SAT_art, 2), round(Q3_acc, 2), name, sep=', ')
```
