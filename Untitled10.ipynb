```python
for i in range(1, 10):
    temp = []
    print(f'Очень счастливые числа в диапазоне {i*1000} : {(i+1)*1000}:')
    print()
    for j in range(i*1000, (i + 1)*1000):
        num = str(j)
        if int(num[0]) + int(num[1]) == int(num[2]) + int(num[3]):
            temp.append(j)
    print(*temp, sep='\t')
    print()
```

```python
import csv
from datetime import datetime
from collections import namedtuple
with open('meetings.csv', 'rt', encoding='utf-8') as file:
    data = list(csv.DictReader(file))
Friend = namedtuple('Friend', ['name', 'time'])
data1 = [Friend(x['surname']+' '+x['name'], 
        datetime.strptime(x['meeting_date']+' '+x['meeting_time'], '%d.%m.%Y %H:%M')) for x in data]
data2 = [x.name for x in sorted(data1, key=lambda x: x.time)]
print(*data2, sep='\n')
```
